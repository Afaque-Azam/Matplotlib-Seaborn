```python
import matplotlib.pyplot as plt
import numpy as np

ages = [
    34, 28, 36, 55, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 52, 24, 19, 53, 50, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 69, 69, 58, 17, 32, 10, 16, 31,
    67, 31, 58, 26, 15, 52, 25, 60, 33, 12,
    3, 26, 23, 36, 40, 39, 21, 66, 53, 39,
    25, 28, 18, 18, 38, 43, 29, 60, 33, 53,
    33, 45, 69, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 52, 26, 26, 15, 32, 32, 30, 27
]

bins = [10, 20, 30, 40, 50, 60, 70]

plt.hist(ages, bins = bins, edgecolor = "black")

plt.axvline(np.mean(ages), color="red", linestyle = "--", linewidth = 2, label = "Average Age")


plt.title("Age distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of viewers")
plt.legend()
# plt.grid(True)
plt.show()
```
