### Sachin Tendulkar's run over time

```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
plt.plot(years,runs)
plt.xlabel("Years")
plt.ylabel("Run Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

### performance comparison

```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label = "Virat Kohli")
plt.plot(years, sehwag, label = "Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
tendulkar =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]


plt.plot(years, kohli, color = 'orange', linestyle = '--', label = "Virat Kohli")
plt.plot(years, sehwag, color = 'green', linestyle = ':', label = "Virender Sehwag")
plt.plot(years, tendulkar, color = 'blue', label = "Sachin Tendulkar")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```
