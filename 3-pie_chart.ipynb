```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

players = ["Sachin", "Sehwag", "Kohli", "Dhoni"]
runs = [18009, 9000,12345,10654]

plt.pie(runs, labels=players)
plt.title("Runs of Indian batsman")
```

```python
# Use custom colors

colors = ["red", "green", "yellow", "blue"]
plt.pie(runs, labels=players, colors=colors)
plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
plt.pie(
    runs,
    labels=players,
    colors=colors,
    wedgeprops={'edgecolor':'black', 'linewidth':2, 'linestyle':'--'}
)
plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
# highlight a slice using explode

explode = [0.1, 0, 0, 0]

plt.pie(
    runs,
    labels=players,
    explode=explode,
    colors=colors
)

plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
plt.pie(
    runs,
    labels=players,
    explode=explode,
    colors=colors,
    shadow=True
)

plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
plt.pie(
    runs,
    labels=players,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=90
)

plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
plt.pie(
    runs,
    labels=players,
    explode=explode,
    colors=colors,
    shadow=True,
    autopct='%1.1f%%'
)

plt.title("Runs of Indian batsman")
plt.axis("equal")
plt.show()
```

```python
# basic pie chart

import matplotlib.pyplot as plt

subjects = ['Math', "Science", "English", "S.St", "Hindi"]
marks = [96,76,87,70,89]

plt.pie(marks, labels = subjects)
plt.title("Marks Distribution")
plt.axis("equal")
plt.show()
```

```python
explode = [0.1,0,0,0,0]

plt.pie(
    marks,
    labels = subjects,
    explode = explode,
    shadow = True,
    startangle = 90,
    autopct = "%1.1f%%"
)
plt.title("Marks Distribution")
plt.axis("equal")
plt.show()
```
