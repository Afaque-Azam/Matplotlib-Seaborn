```python
import matplotlib.pyplot as plt
 
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'TV', 'Sleeping']

plt.figure(figsize=(10,6))

plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, alpha=0.8)

plt.legend(loc='upper left')
plt.title("Weekly Activity Tracker")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
```

### With % Contribution (Normalized)

```python
import matplotlib.pyplot as plt
import numpy as np

days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Sleeping']

total = np.array(studying) + np.array(playing) + np.array(sleeping)

plt.stackplot(
    days,
    np.array(studying)/total,
    np.array(playing)/total,
    np.array(sleeping)/total,
    labels = labels
)

plt.legend(loc='upper left')
plt.title("Weekly Activity Tracker")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
```
