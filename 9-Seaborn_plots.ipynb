## Count Plot  (categorical data)

```python
sns.countplot(x= 'sex', data=tips)
plt.title("Total count By Sex")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x= 'sex', data=tips, hue="smoker")
plt.title("Total smoker count By Sex")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'day', data=tips)
plt.title("Total_bill Per Day")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'day', data=tips, hue="smoker")
plt.title("Total Smoker Per Day")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'day', data=tips, hue="sex")
plt.title("Total Sex Per Day")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'time', data=tips)
plt.title("Total_bill Per Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'time', data=tips, hue='smoker')
plt.title("Total_bill Per Smoker")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'time', data=tips, hue='sex')
plt.title("Total_bill(by sex) Per Day")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'size', data = tips)
plt.title("Total_bill Per Family Size")
plt.xlabel("Family Size")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x = 'size', data = tips, hue='smoker')
plt.title("Total_bill Per Family Size")
plt.xlabel("Family Size")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Histplot (numerical data)

```python
tips["total_bill"].max()
```

```python
tips.nlargest(5, 'total_bill')
```

```python
bins = [0,5,10,15,20,25,30,35,40,45,50,55,60]
sns.histplot(x="total_bill", bins = bins, data=tips, kde=True)
plt.title("Total_bill distribution")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
tips.nlargest(5, 'tip')
```

```python

sns.histplot(x="tip", data=tips, kde=True)
plt.title("Total_tip distribution")
plt.tight_layout()
plt.grid(True)
plt.show()
```

## Lineplot, Scatterplot (num vs num data)

```python
sns.lineplot(x = "total_bill", y = "tip", data = tips, hue = "time")
plt.title("Line Plot Example")
plt.show()
plt.tight_layout()
```

### Scatter plot

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "time")
plt.title("Scatter Plot Example")
plt.show()
plt.tight_layout()
```

### Bar plot  (cat vs num data)

```python
sns.barplot(x = "day", y = "total_bill", data = tips, estimator=np.size, hue='sex')
plt.title("Bar Plot Example")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(x = "day", y = "total_bill", data = tips, estimator=np.size)
plt.title("Bar Plot Example")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Box plot (cat vs num)

```python
sns.boxplot(x = "day", y = "total_bill", data = tips)
plt.title("Box Plot Example")
plt.show()
```

## violinplot (cat vs num)

```python
sns.violinplot(x = 'day', y='total_bill', data=tips)
plt.title("Violin Plot Example")
plt.show()
```

## stripplot (cat vs num)

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.title("Strip Plot Example")
plt.show()
```

## swarmplot (cat vs num)

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.title("Swarm Plot Example")
plt.show()
```

## Regression plot

```python
sns.regplot(x = 'total_bill', y = 'tip', data=tips)
plt.title("Relation between Total_bill & Tip")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Heatmap

```python
tipscorr = tips[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True)
plt.title("Correlation Between total_bill, tip & size")
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
pv = flights.pivot(index = "month", columns = 'year', values = "passengers")
pv
```

```python
sns.heatmap(pv, annot = True, fmt='d')
plt.title("heatmap of Passengers")
plt.show()
```

### working with the pandas DataFrame

```python
import pandas as pd

df = pd.DataFrame({
     "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.barplot(x = "age", y = "salary", hue = "gender", data = df)
plt.title("Salary vs Age Bar Plot")
plt.show()
```

```python
sns.scatterplot(x = "age", y = "salary", hue = "gender", data = df)
plt.title("Salary vs Age Scatter Plot")
plt.show()
```
